## **BAGIAN 2: Penggabungan Tabel dari Relasi Kolom**

Sebelum kita memulai untuk melangkah lebih lanjut, kita akan menggunakan dua tabel yakni tabel orders dan order_details. Mari kita melihat isinya terlebih dahulu, yuk!

````python
# Ketahui isi tabel orders
sql(
"""
SELECT *
FROM orders
LIMIT 5
"""
)
````

````python
# Ketahui isi tabel order_details
sql(
"""
SELECT *
FROM order_details
LIMIT 5
"""
)
````

Manakah dari kedua tabel tersebut yang memiliki kecocokan untuk dijadikan penghubung yang dapat menjadi **key_column**? Tunjukan isi yang sama di kedua tabel



```sql
SELECT column_name1 FROM table_name1
JOIN table_name2 ON table_name1.key = table_name2.key
```



````python
# Silahkan isi bagian yang kosong
sql(
"""
SELECT * FROM order_details
JOIN orders ON order_details.[1]____ = orders.[2]____
LIMIT 5;
"""
)
````

Ada langkah mudah untuk menghubungkan key_columns dengan cara berikut ini (outputnya Inner Join):

```sql
SELECT * FROM table_name1, table_name2
WHERE unique_key1=unique_key2;
```

Operator (,) untuk melakukan join sedangkan operator (=) untuk menghubungkan satu sama lain. Ini hanya menampilkan data yang cocok saja ya! Proses ini hampir mirip dengan melakukan VLOOKUP lho! 

Pengaruh tidak ya, apabila table_name2 dulu baru table_name1?

````python
# Silahkan isi bagian yang kosong
sql(
"""
SELECT * FROM [1]____, [2]____
WHERE [3]____=[4]____
LIMIT 5;
""")
````

Lalu bagaimana apabila **where** dihapuskan? Ini akan menjadi cross join, artinya setiap baris data pada kedua tabel akan dihubungkan satu sama lain - tanpa ada hubungan. Atau sederhananya row_baris_1 * row_baris_2 akan menjadi semua hasil row. 

**Cross join** terjadi dari penggabungan tabel tanpa kondisi, dan menghasilkan seluruh penggabungan data seperti proses perkalian.

````python
sql(
"""
SELECT orders.*, order_details.*  FROM orders, order_details
LIMIT 5;
""")
````
