## **BAGIAN 1: Apa itu python?**

Python merupakan bahasa pemrograman yang dibentuk oleh **Guido van Rossum** pada tahun 1991. Python dapat dikatakan sebagai programming language dengan perkembangan yang cepat karena bersifat **open-source** dan **general-purpose programming**. Python merupakan salah satu programming language yang cocok untuk **scripting** dan bisa berjalan dalam berbagai **platform OS dan IoT** yang mana tergolong ke dalam **high level programming language**, python sangat mudah dimengerti karena syntax-nya yang sederhana karena memiliki kesamaan dengan bahasa inggris. 

Kegunaan python pun sangatlah luas diantaranya:
1. **Data Science** ini yang akan kita fokuskan. Hal ini didorong adanya library dalam kepentingan analisis dan membuat model prediktif dalam data science.
2. Web-development
3. Game-development
4. Computer Vision
5. dan sebagainya

Adapun kelebihan dengan menggunakan python diantaranya:
1. Python dirancang untuk **mudah ditulis dan dibaca**, serta memiliki beberapa kesamaan dengan bahasa Inggris dengan pengaruh dari matematika.
2. Python menggunakan **baris baru untuk mengakhiri perintah**, dibandingkan dengan bahasa pemrograman lain yang sering menggunakan titik koma atau tanda kurung.
3. Python **bergantung pada indentasi**, menggunakan spasi, untuk mendefinisikan ruang lingkup; seperti lingkup loop, fungsi, dan kelas. Bahasa pemrograman lainnya sering menggunakan kurung kurawal untuk tujuan ini
4. Jika kalian mencari di internet tentang programming python, biasanya akan ada 2 jenis python, python 2 dan python 3. Versi utama Python yang paling baru adalah Python 3. Saat ini, **python 2** sudah memasuki end-of-life pada **Januari 2020**.

List Library yang umum digunakan **data science**, silahkan klik yang berwarna biru:

1. **[Numpy](https://numpy.org/doc/) (numerical python)** merupakan library yang memudahkan dalam pendefinisian array (1D, 2D, 3D atau nD) dan digunakan untuk komputasi aljabar linier.
2. **[Scipy](https://docs.scipy.org/doc/scipy/) (Scientific Python)** merupakan library yang melengkapi numpy untuk keperluan komputasi saintifik lanjut seperti aljabar linier, integrasi dan diferensiasi numerik, transformasi Fourier, optimasi, interpolasi, statistik dan yang lainnya.
3. **[Pandas](https://pandas.pydata.org/docs/) (Python Data Analysis Library)** adalah library untuk pengolahan data dalam bentuk tabular (seperti excel) yang merupakan de facto library (pustaka yang umum digunakan dalam prakteknya) bagi data scientist dalam mengolah data dari berbagai sumber seperti file CSV, TSV, Excel, SQL queries, Google BigQuery, SAS, Stata, SPSS, dsb.
4. **[Matplotlib](https://matplotlib.org/)** digunakan untuk visualisasi dari data ke dalam berbagai bentuk grafik 2D atau 3D, seperti line chart, bar chart, histogram, polar chart, error bar plot, dan jenis grafik lainnya.
5. **[Scikit-learn](https://scikit-learn.org/)** adalah Scipy Toolkit yang ditujukan untuk menghasilkan model prediktif dengan menggunakan machine learning.
6. **[Seaborn](https://seaborn.pydata.org/)** merupakan library yang dibuat dari matplotlib yang ditujukan oleh visualisasi grafik statistik dengan warna yang menawan, dan terintegrasi dengan baik dengan pandas.
7. Dan terus berkembang hingga saat ini.

### **Aturan Variabel**

Variabel dikenal sebagai **elemen definisi** yang dapat digunakan untuk **reproducibility** (bisa digunakan berulang-ulang) dengan ditandai = (assignment).

```python
# Konsep Reproducibility
# RUMUS BMI = Weight (kg)/ Height(m2)^2
# Tanpa variabel
print(20/(1.20**2))
# Dengan variabel
weight = 20
height = 1.2
BMI = weight/(height**2)
print(BMI)
```

Dengan aturan deklarasi di Python. 

1. Nama dari sebuah variabel harus **dimulai** dengan **huruf (a-z, A-Z)** atau karakter **garis bawah/underscore (_)** dan **tidak dapat dimulai dengan angka (0-9).**
2. Variabel hanya boleh mengandung karakter alfabet dan bilangan dan underscore **(a-z, A-Z, 0-9, _)**
3. Variabel bersifat **case-sensitive** yang mengartikan bahwa penulisan variabel **TINGGI**, **tinggi**, dan **Tinggi** merujuk pada tiga variabel berbeda.

contoh

```python
angka1 = 10
Angka_2 = 40
Pembuka = "Halo, apakabar?"
angka1, Angka_2 = 10, 40
salam = "Selamat Pagi di Bogor"; Penutup = "Sampai Jumpa"
```
